## Line Chart of the Top 10 Stocks

```python
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt

def fetch_stock_data(ticker, start_date, end_date):
    try:
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        if stock_data.empty:
            raise ValueError(f"No data found for ticker: {ticker}")
        return stock_data['Adj Close']  # Select only the 'Adj Close' column
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return pd.Series()

def plot_univariate_analysis(data, ticker):
    plt.figure(figsize=(12, 6))
    plt.plot(data, label=f'{ticker} Adj Close', color='blue')

    # Highlighting the high and low peaks
    if not data.empty:
        high_peak = data.max()
        high_peak_date = data.idxmax()
        low_peak = data.min()
        low_peak_date = data.idxmin()

        plt.scatter(high_peak_date, high_peak, color='green', label='High Peak', zorder=5)
        plt.scatter(low_peak_date, low_peak, color='red', label='Low Peak', zorder=5)

        plt.title(f'Adjusted Close Prices for {ticker}')
        plt.xlabel('Date')
        plt.ylabel('Adjusted Close Price')
        plt.legend()
        plt.grid(True)
        plt.show()

        return high_peak_date, high_peak, low_peak_date, low_peak
    else:
        return None, None, None, None

if __name__ == "__main__":
    tickers = ['AMZN', 'SNCR','DOMO','UBER', 'BA','ABNB', 'NAUT', 'VITL','WDAY', 'DNLI']
    start_date = '2021-01-01'
    end_date = '2023-01-01'

    for ticker in tickers:
        print(f"\nAnalyzing {ticker}...")
        adj_close_data = fetch_stock_data(ticker, start_date, end_date)

        # Plot data and get peaks
        high_peak_date, high_peak, low_peak_date, low_peak = plot_univariate_analysis(adj_close_data, ticker)

        if high_peak_date and low_peak_date:
            # Print insights
            print(f'{ticker} - Highest adjusted close price: {high_peak} on {high_peak_date}')
            print(f'{ticker} - Lowest adjusted close price: {low_peak} on {low_peak_date}')
        else:
            print(f"Data not available for {ticker}.")
```

## Daily Returns vs Standard Deviation

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Function to fetch stock data
def fetch_stock_data(ticker, start_date, end_date):
    try:
        stock_data = yf.download(ticker, start=start_date, end=end_date)
        if stock_data.empty:
            raise ValueError(f"No data found for ticker: {ticker}")
        return stock_data['Adj Close']  # Select only the 'Adj Close' column
    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")
        return pd.Series()

# Fetch stock data for a list of tickers
def fetch_all_data(tickers, start_date, end_date):
    all_data = pd.DataFrame()
    for ticker in tickers:
        data = fetch_stock_data(ticker, start_date, end_date)
        if not data.empty:
            all_data[ticker] = data
    return all_data

# Function to calculate and plot returns
def calculate_and_plot_returns(df):
    # Calculate daily returns
    returns_df = df.pct_change().dropna()  # Drop the first row with NaN values

    # Display the first 5 rows of the returns DataFrame
    print("Daily Returns DataFrame:")
    print(returns_df.head(5))

    # Calculate the standard deviation of returns
    Std_Dev_Return = pd.DataFrame(np.std(returns_df), columns=["Std_Dev_Return"])

    # Print the standard deviation of returns
    print("\nStandard Deviation of Returns:")
    print(Std_Dev_Return)

    # Calculate the mean returns
    mean_returns = returns_df.mean()

    # Plot the data
    plt.figure(figsize=(10, 6))

    # Scatter plot of standard deviation vs mean returns
    plt.scatter(Std_Dev_Return, mean_returns, c=mean_returns, cmap='viridis', marker='o')
    plt.colorbar(label='Mean Daily Returns')

    for i, ticker in enumerate(returns_df.columns):
        plt.annotate(ticker, (Std_Dev_Return.iloc[i, 0], mean_returns[i]))

    plt.title('Mean Returns vs Standard Deviation of Returns')
    plt.xlabel('Standard Deviation (Risk)')
    plt.ylabel('Mean Daily Return')
    plt.grid(True)
    plt.show()

if __name__ == "__main__":
    # List of tickers based on hypothetical Trump's top 10 stocks
    tickers = ['AMZN', 'SNCR','DOMO','UBER', 'BA','ABNB', 'NAUT', 'VITL','WDAY', 'DNLI']
    end_date = '2023-01-01'

    # Fetch all stock data
    df = fetch_all_data(tickers, start_date, end_date)

    # Calculate and plot returns
    calculate_and_plot_returns(df)
```

When we look at our average daily returns and compare them to their standard deviations, we can see that BA is the only stock holding a positive average daily return with one of the lowest standard deviations or in this case the lowest risk and volatility.

Most of our stocks hold relatively low standard deviations. However, their daily average returns are very low (negative).

This suggests that on a daily basis, Jeff Bezos' investments perform very poorly at the exception of BA. Still, BA has a very low average positive daily return.

##Correlation Matrix

What better way to start our bivariate analysis than building a correlation matrix that will help us identify strong and weak correlations between our different stocks?

```python
import yfinance as yf
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Create an empty DataFrame
df = pd.DataFrame()

def fetch_stock_data(ticker, start_date, end_date):
    stock_data = yf.download(ticker, start=start_date, end=end_date)
    return stock_data['Adj Close']  # Select only the 'Adj Close' column

def main():
    # List of stock tickers
    stocks =  ['AMZN', 'SNCR','DOMO','UBER', 'BA','ABNB', 'NAUT', 'VITL','WDAY', 'DNLI']
    # Date range
    start_date = '2021-01-01'
    end_date = '2023-01-01'

    # Fetch data for each stock and add to the DataFrame
    for stock in stocks:
        df[stock] = fetch_stock_data(stock, start_date, end_date)

    # Construct the correlation matrix
    correlation_matrix = df.corr()

    # Create a heatmap
    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('Correlation Matrix Heatmap')
    plt.show()

# Run the main function if this script is executed
if __name__ == "__main__":
    main()
```

Note: This heatmap above presents us with the different correlations for our adjusted closing prices.

## Define Decision Variables

```python
# Declare decision variables
m.AMZN = Var(within=NonNegativeReals, bounds=(0, 1))
m.SNCR = Var(within=NonNegativeReals, bounds=(0, 1))
m.DOMO = Var(within=NonNegativeReals, bounds=(0, 1))
m.UBER = Var(within=NonNegativeReals, bounds=(0, 1))
m.BA = Var(within=NonNegativeReals, bounds=(0, 1))
m.ABNB = Var(within=NonNegativeReals, bounds=(0, 1))
m.NAUT = Var(within=NonNegativeReals, bounds=(0, 1))
m.VITL = Var(within=NonNegativeReals, bounds=(0, 1))
m.WDAY = Var(within=NonNegativeReals, bounds=(0, 1))
m.DNLI = Var(within=NonNegativeReals, bounds=(0, 1))

```

These variables will be used in the optimization model to allocate the portfolio across these stocks. The goal is to determine the optimal proportion of the portfolio to invest in each stock to achieve a specific objective, such as maximizing returns or minimizing risks, while respecting constraints.

- The bounds of (0,1)  means that we want to understand that for each stock we can allocate between 0% and 100% of the portfolio , but not more or less.

## Define Objective Function

```python
m.objective = Objective(expr = df_return[0]*m.AMZN +
                        df_return[1]*m.SNCR +
                        df_return[2]*m.DOMO +
                        df_return[3]*m.UBER +
                        df_return[4]*m.BA +
                        df_return[5]*m.ABNB +
                        df_return[6]*m.NAUT +
                        df_return[7]*m.VITL +
                        df_return[8]*m.WDAY +
                        df_return[9]*m.DNLI,
                        sense=maximize)
```

- The objective function m.objective is designed to maximize the total return of the portfolio by adjusting the investment proportions in each stock. It does this by calculating the weighted sum of the returns for each stock, where the weights are the proportions of the total portfolio invested in each stock. The maximize sense indicates that the optimization process will seek to find the allocation that yields the highest possible return based on the given return values for each stock.

## Define Return

```python
#return
m.return_floor = Constraint(expr = m.objective >= 0.0009)

#the stock portfolio should add up to 100
m.sum_proportions = Constraint(expr = m.AMZN + m.SNCR + m.DOMO + m.UBER + m.BA + m.ABNB + m.NAUT + m.VITL + m.WDAY + m.DNLI == 1)

#set maximum risk
m.total_risk = Constraint(expr= m.AMZN + m.SNCR + m.DOMO + m.UBER + m.BA + m.ABNB + m.NAUT + m.VITL + m.WDAY + m.DNLI >= 0)
```

- This constraint ensures that the total return of the portfolio is at least 0.0009 (or 0.09%).This constraint ensures that the sum of all investment proportions in the portfolio equals 1 (or 100%).This constraint specifies that the sum of the proportions of all stocks should be greater than or equal to 0.

## Define Risk

```python
# creating calculations table for calculate the risk
# 3. Calculate risk
def calc_risk(m):
  variables = m.AMZN, m.SNCR, m.DOMO, m.UBER, m.BA, m.ABNB, m.NAUT, m.VITL, m.WDAY, m.DNLI
  tickers = ['AMZN', 'SNCR', 'DOMO', 'UBER', 'BA', 'ABNB', 'NAUT', 'VITL', 'WDAY', 'DNLI']
  risk_exp = 0
  for i in range(len(variables)):
    for j in range(len(variables)):
      risk_exp += variables[i]*df_cov.at[tickers[i],tickers[j]]*variables[j]
  return risk_exp

# We are going to use this expression to compute the risk
expr_risk = calc_risk(m)

# 3. Max risk
max_risk =  0.01
# Sequence of risk levels
risk_limits = np.arange(0.0003, max_risk, 0.0005) # take tiny steps
risk_limits
```

- This function computes the total risks of the portfolio based on the covariance matrix of stock returns. Using risk limits to evaluate different risk-return profiles. ( 0.0003 and 0.0005)
- This outputs the return of 0.09 % and the risks of 1% allocation to the stocks determine the total risk and the total return.

## Define Solution - Solver 1

```python
 #Uupdating risk contraint for each limit and then solving the problem
param_analysis = {} # key=risk, value =stock allocations
returns = {} # key=risk, value = return
for r in risk_limits:
  # WE REMOVE AND RECALCULATE THE RISK IN EACH ITERATION
  m.del_component(m.total_risk)
  # The LHS remains unchanged; we only modify the RHS (risk threshold)
  m.total_risk = Constraint(expr = expr_risk <= r)
  # run solver
  result = SolverFactory('ipopt', executable=ipopt_executable).solve(m).write()
  # store our allocation proportions
  param_analysis[r] = [m.AMZN(), m.SNCR(), m.DOMO(), m.UBER(),m.BA(), m.ABNB(), m.NAUT(),m.VITL(),m.WDAY(), m.DNLI()]
  # store our returns
  # returns[r] = m.objective()
  returns[r] =  m.AMZN()*df_return[0] + m.SNCR()*df_return[1] + m.DOMO()*df_return[2] + m.UBER()*df_return[3] + m.BA()*df_return[4] + m.ABNB()*df_return[5] + m.NAUT()*df_return[6] + m.VITL()*df_return[7] + m.WDAY()*df_return[8] + m.DNLI()*df_return[9]
```

# Model 2 ( Minimization)

```python
# Import the minimize function
from scipy.optimize import minimize

# Define the objective function: Portfolio variance
def portfolio_variance(weights, df_cov):
    return weights.T @ df_cov @ weights

# Constraints: Sum of weights = 1
constraints = ({'type': 'eq', 'fun': lambda weights: np.sum(weights) - 1})

# Bounds: Weights should be between 0 and 1
bounds = tuple((0, 1) for stock in range(len(df.columns)))

# Initial guess: Equal distribution
initial_weights = np.array([1/len(df.columns)] * len(df.columns))

# Minimize the portfolio variance
result = minimize(portfolio_variance, initial_weights, args=(df_cov,), method='SLSQP', bounds=bounds, constraints=constraints)

optimal_weights = result.x
print("Optimal Weights:", optimal_weights)
```

## Optimal Weights for Minimizing Portfolio Return
